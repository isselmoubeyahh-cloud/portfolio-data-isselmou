```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the DataFrame
data = {
    'marital_statut': ['Célibataire', 'marié(e)', 'veuf/veuve', 'divorcé(e)'],
    'Non': [98.0, 89.3, 81.0, 89.4],
    'Oui': [2.0, 10.7, 19.0, 10.6]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame for seaborn
df_melted = df.melt(id_vars='marital_statut', value_vars=['Non', 'Oui'], var_name='Hypertension', value_name='Pourcentage')

# Set the style
sns.set(style="whitegrid")

# Create the bar plot with a red palette with different shades of red
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='marital_statut', y='Pourcentage', hue='Hypertension', data=df_melted, palette=sns.color_palette("Reds", n_colors=2))

# Add percentage labels on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'baseline',
                fontsize=10, color='black', xytext = (0, 5),
                textcoords = 'offset points')

# Add titles and labels

ax.set_xlabel('Marital Status', fontsize=12)
ax.set_ylabel('Pourcentage', fontsize=12)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Données d'exemple correspondant à votre image
data = {
    'Réponse': ['Non', 'Non', 'Oui', 'Oui'],
    'Milieu de Résidence': ['Urbain', 'Rural', 'Urbain', 'Rural'],
    'Pourcentage': [93.2, 91.2, 6.8, 8.8]
}

df = pd.DataFrame(data)

# Créer un diagramme à barres empilées
fig, ax = plt.subplots(figsize=(10, 6))

# Positions des barres sur l'axe x
bar_width = 0.6

# Tracer les barres pour 'Non'
bars_non = df[df['Réponse'] == 'Non']
bars1 = ax.bar('Non', bars_non[bars_non['Milieu de Résidence'] == 'Urbain']['Pourcentage'], 
               bar_width, label='Urbain', color='#820000')
bars2 = ax.bar('Non', bars_non[bars_non['Milieu de Résidence'] == 'Rural']['Pourcentage'], 
               bar_width, bottom=bars_non[bars_non['Milieu de Résidence'] == 'Urbain']['Pourcentage'], label='Rural', color='#ED8E8E')

# Tracer les barres pour 'Oui'
bars_oui = df[df['Réponse'] == 'Oui']
bars3 = ax.bar('Oui', bars_oui[bars_oui['Milieu de Résidence'] == 'Urbain']['Pourcentage'], 
               bar_width, label='Urbain', color='#820000')
bars4 = ax.bar('Oui', bars_oui[bars_oui['Milieu de Résidence'] == 'Rural']['Pourcentage'], 
               bar_width, bottom=bars_oui[bars_oui['Milieu de Résidence'] == 'Urbain']['Pourcentage'], label='Rural', color='#ED8E8E')

# Ajouter les pourcentages au-dessus des barres
for bar in bars1 + bars2 + bars3 + bars4:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2, f'{height}%', ha='center', va='center', color='black', fontsize=12)

# Étiquettes et titre
ax.set_xlabel('Réponse')
ax.set_ylabel('Pourcentage')
ax.set_title('Avez-vous une pression artérielle élevée ou de l\'hypertension')
ax.legend(['Urbain', 'Rural'])

plt.show()
```
